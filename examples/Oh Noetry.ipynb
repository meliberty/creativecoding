```python
import pronouncing
from textblob import TextBlob 

inpoem = 'Roses are red, violets are blue\nsugar is sweet and so are you'

blob = TextBlob(inpoem)

for word,pos in blob.tags:
    out = ""
    out += " " + word
    rhymes = pronouncing.rhymes(word)

    for r in rhymes:
        rbl = TextBlob(r)
        if rbl.tags[0][1] == pos:
            out += " " + rbl.tags[0][0]
    print(out)
```

```python
tens = []
for s in blob.sentences:
    syll = 0
    blank = 0
    for word in s.words:
        phones = pronouncing.phones_for_word(word)
        if (len(phones)> 0):
            syll_count = pronouncing.syllable_count( phones[0] )
            syll += syll_count
        else:
            blank += 1
            
    if (blank < 1 and syll in [4,8,12,16,20,24]):
        tens.append(s)
tens     
```

```python
with open('frankenstein.txt', 'r') as file:
    text = file.read()
    
flob = TextBlob(text)
np = flob.noun_phrases
```

```python
def sum_syllables(phrase):
    wordlist = phrase.split()
    fail = 0
    syllable_sum = 0
    for word in wordlist:
        phones = pronouncing.phones_for_word(word)
        if (len(phones) > 0):
            syllables = pronouncing.syllable_count(phones[0])
            syllable_sum += syllables
        else:
            fail += 1
    if (fail > 0):
        return 0
    else:
        return syllable_sum

    
import random

lines = []

for phrase in np:
    count = sum_syllables(phrase)
    if (count is 6):
        lines.append(phrase)
        
def find_pair(lines):
    # pick a random line
    line1 = random.choice(lines)
    
    line2 = ''
    # look for a match
    for li in lines:
        
        if (li.split()[:1] in pronouncing.rhymes(line1.split()[:1])):
            line2 = li
    return line1 + "\n" + line2
        
print(find_pair(lines))
        
```
